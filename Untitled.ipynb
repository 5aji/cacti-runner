```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
data = pd.read_csv("output.csv")
import seaborn as sns
sns.set_theme(style="darkgrid")
mpl.rcParams['figure.dpi'] = 300
```

```python
data.columns
```

```python
plt.figure(dpi=200)
sns.relplot(data=data[data["associativity"] != 0], kind='line',x="n_entries", y=" Random cycle time (ns)", hue="associativity")
plt.show()
```

```python
sns.relplot(data=data, kind='line',x="n_entries", y=" Dynamic read energy (nJ)", style="associativity", hue="associativity", )
plt.show()
```

```python
sns.relplot(data=data, kind='line',x="n_entries", y=" Standby leakage per bank(mW)" , hue="associativity")
plt.show()
```

```python
sns.relplot(data=data, kind='line',x="n_entries", y=" Area (mm2)" , hue="associativity")
plt.show()
```

```python
g = sns.FacetGrid(data, col="addr_size", hue="associativity")
g.map(sns.lineplot, "n_entries", " Access time (ns)")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data, col="addr_size", hue="associativity")
g.map(sns.lineplot, "n_entries", " Standby leakage per bank(mW)")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data, col="addr_size", hue="associativity")
g.map(sns.lineplot, "n_entries", " Dynamic read energy (nJ)")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data, col="addr_size", hue="associativity")
g.map(sns.lineplot, "n_entries", " Area (mm2)")
g.add_legend()
plt.show()
```
